```python
from scipy import signal
```

```python
amp = 2 * np.sqrt(2)
f, t, Zxx = signal.stft(pureDataSample, sample_rate, nperseg=1000)
plt.pcolormesh(t, f, np.abs(Zxx), vmin=0, vmax=amp, shading='gouraud')
plt.title('STFT Magnitude')
plt.ylabel('Frequency [Hz]')
plt.xlabel('Time [sec]')
```
